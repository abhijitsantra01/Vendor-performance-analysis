```python
from collections import defaultdict
import json

# Suppose you already loaded your JSON file into a variable called 'data'
# Example: data = json.load(open('news.json'))

clusters = defaultdict(list)

for item in data['vendor_data']:
    company = item['company_name']
    event = item['news_event']
    clusters[company].append(event)

# Convert defaultdict back to normal dict (optional)
clusters = dict(clusters)

# Now you can print or inspect it
print(json.dumps(clusters, indent=2))
```

```python
import openai

openai.api_key = "sk-proj-9R02L3hC1J80Jl9KKqhk_*********"  

story_reports = {}

for company, events in clusters.items():
    prompt = f"""
    You are a financial analyst writing a concise business sentiment and trend report.

    Company: {company}
    News events:
    {chr(10).join(events)}

    Based on these events, write:
    1. A short summary of major developments.
    2. The overall sentiment (positive/neutral/negative) and why.
    3. A forward-looking statement on business outlook.
    """

    response = openai.chat.completions.create(
        model="gpt-4o-mini",  # or "gpt-4o" / your available model
        messages=[{"role": "user", "content": prompt}],
        temperature=0.6
    )

    story_reports[company] = response.choices[0].message.content
```

```python
# ---- existing story generation loop ----
story_reports = {}

for company, events in clusters.items():
    prompt = f"""
    You are a financial analyst writing a concise business sentiment and trend report.

    Company: {company}
    News events:
    {chr(10).join(events)}

    Based on these events, write:
    1. A short summary of major developments.
    2. The overall sentiment (positive/neutral/negative) and why.
    3. A forward-looking statement on business outlook.
    """

    response = openai.chat.completions.create(
        model="gpt-4o-mini",  # compact, fast, cost-efficient model
        messages=[{"role": "user", "content": prompt}],
        temperature=0.6
    )

    story_reports[company] = response.choices[0].message.content

# ---- new section: investment relationship analysis ----
investor_company = "Alphabet (Google)"
target_vendor = "Reliance Industries"

# Get the company’s story report as context (optional but powerful)
company_report = story_reports.get(target_vendor, "No report available.")

# Build the new prompt
investment_prompt = f"""
You are an experienced investment strategist.

Investor company: {investor_company}
Target vendor: {target_vendor}

Below is a recent business sentiment and trend report for {target_vendor}:

{company_report}

Based on this information, analyze and answer:
1. What is the current business and strategic outlook for {target_vendor}?
2. What are the key risks and opportunities for {investor_company} as a partner or investor?
3. Should {investor_company} continue, expand, or reconsider its relationship with {target_vendor}, and why?
Provide your response as a short, well-structured investment advisory note.
"""

# Make the LLM call
investment_response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a senior financial analyst providing objective investment guidance."},
        {"role": "user", "content": investment_prompt}
    ],
    temperature=0.6
)

# Print the AI-generated investment recommendation
investment_analysis = investment_response.choices[0].message.content
print("\n===== INVESTMENT RECOMMENDATION REPORT =====\n")
print(investment_analysis)
```
